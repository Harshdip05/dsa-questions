# 🔹 Problem summary

For each number:

* Positive → `1`
* Negative → `2`
* Zero → `0`

---

# ✅ Solution 1: Simple loop (in-place)

```python
N = int(input())
A = list(map(int, input().split()))

for i in range(N):
    if A[i] > 0:
        A[i] = 1
    elif A[i] < 0:
        A[i] = 2
    else:
        A[i] = 0

print(*A)
```

### 🔸 Explanation

* Traverse the array once
* Modify each element directly

### ⏱ Time Complexity

* **O(N)** → one pass through the list

### 💾 Space Complexity

* **O(1)** → no extra space (in-place)

---

# ✅ Solution 2: List comprehension (Pythonic)

```python
N = int(input())
A = list(map(int, input().split()))

A = [1 if x > 0 else 2 if x < 0 else 0 for x in A]

print(*A)
```

### 🔸 Explanation

* Creates a new list using a compact expression

### ⏱ Time Complexity

* **O(N)**

### 💾 Space Complexity

* **O(N)** → new list is created

---

# ✅ Solution 3: Using a helper function

```python
def transform(x):
    if x > 0:
        return 1
    elif x < 0:
        return 2
    return 0

N = int(input())
A = list(map(int, input().split()))

A = list(map(transform, A))

print(*A)
```

### 🔸 Explanation

* `map()` applies a function to every element

### ⏱ Time Complexity

* **O(N)**

### 💾 Space Complexity

* **O(N)**

---

# ✅ Solution 4: Using separate result list

```python
N = int(input())
A = list(map(int, input().split()))

result = []
for x in A:
    if x > 0:
        result.append(1)
    elif x < 0:
        result.append(2)
    else:
        result.append(0)

print(*result)
```

### 🔸 Explanation

* Builds a new list step by step

### ⏱ Time Complexity

* **O(N)**

### 💾 Space Complexity

* **O(N)**

---

# ✅ Solution 5: Using math trick (advanced but neat)

```python
N = int(input())
A = list(map(int, input().split()))

A = [(x > 0)*1 + (x < 0)*2 for x in A]

print(*A)
```

### 🔸 Explanation

* `(x > 0)` → True/False → 1/0
* `(x < 0)` → True/False → 1/0

So:

* positive → `1 + 0 = 1`
* negative → `0 + 2 = 2`
* zero → `0 + 0 = 0`

### ⏱ Time Complexity

* **O(N)**

### 💾 Space Complexity

* **O(N)**
