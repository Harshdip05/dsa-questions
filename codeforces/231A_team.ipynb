```python
n = int(input())

count = 0
for i in range(n):
    a, b, c = map(int, input().split())
    if (a == 1 and b == 1) or (a==1 and c==1) or (b==1 and c==1):
        count +=1
print(count)        

```

```python
n = int(input())
count = 0

for _ in range(n):
    a, b, c = map(int, input().split())
    if a + b + c >= 2:
        count += 1

print(count)
```
