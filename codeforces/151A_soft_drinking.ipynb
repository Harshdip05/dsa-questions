```python
values = input().split()

n = int(values[0])
k = int(values[1])
l = int(values[2])
c = int(values[3])
d = int(values[4])
p = int(values[5])
nl = int(values[6])
np = int(values[7])

total_drink = k * l
total_slices = c * d

toasts_from_drink = total_drink // nl
toasts_from_lime = total_slices
toasts_from_salt = p // np

total_toasts = min(toasts_from_drink, toasts_from_lime, toasts_from_salt)

print(total_toasts // n)
```

```python
def max_toasts(n, k, l, c, d, p, nl, np):
    drink = (k * l) // nl
    lime = c * d
    salt = p // np
    return min(drink, lime, salt) // n

n, k, l, c, d, p, nl, np = map(int, input().split())

print(max_toasts(n, k, l, c, d, p, nl, np))
```
