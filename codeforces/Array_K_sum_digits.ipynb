This is a straightforward problem, but there are **multiple ways to solve it**, each with slightly different thinking. Let’s go through them clearly.

---

# 🧠 Problem Summary

* Input:

  * `N` → number of digits
  * A string of digits (like `"13305"`)
* Task:
  👉 **Add all digits**

Example:

```
"13305" → 1 + 3 + 3 + 0 + 5 = 12
```

---

# ✅ Method 1: Simple Loop (Best & Most Common)

### 💻 Code

```python
n = int(input())
s = input()

total = 0
for ch in s:
    total += int(ch)

print(total)
```

### 🧠 Idea

* Loop through each character
* Convert to int
* Add

### ⏱️ Complexity

* Time: **O(n)**
* Space: **O(1)**

---

# 🚀 Method 2: Using `sum()` + `map()`

### 💻 Code

```python
n = int(input())
s = input()

print(sum(map(int, s)))
```

### 🧠 Idea

* Convert each character to int using `map`
* Sum all values directly

### ⏱️ Complexity

* Time: **O(n)**
* Space: **O(n)** (temporary map)

---

# ⚡ Method 3: List Comprehension

### 💻 Code

```python
n = int(input())
s = input()

print(sum([int(ch) for ch in s]))
```

### 🧠 Idea

* Create list of digits
* Then sum

### ⏱️ Complexity

* Time: **O(n)**
* Space: **O(n)**

---

# 🔁 Method 4: Without Converting to int (ASCII Trick)

### 💻 Code

```python
n = int(input())
s = input()

total = 0
for ch in s:
    total += ord(ch) - ord('0')

print(total)
```

### 🧠 Idea

* Characters are stored as ASCII
* `'0' → 48`, `'1' → 49`, etc.
  👉 So digit = `ord(ch) - ord('0')`

---

# 🔥 Method 5: Recursive (Not Recommended for large N)

```python
def digit_sum(s):
    if not s:
        return 0
    return int(s[0]) + digit_sum(s[1:])

n = int(input())
s = input()

print(digit_sum(s))
```

### ⚠️ Problem

* Stack overflow for large N (up to 10⁶)

---

# 📊 Comparison

| Method      | Time | Space | Best Use |
| ----------- | ---- | ----- | -------- |
| Loop        | O(n) | O(1)  | ✅ Best   |
| sum + map   | O(n) | O(n)  | Clean    |
| List comp   | O(n) | O(n)  | Pythonic |
| ASCII trick | O(n) | O(1)  | Fast     |
| Recursion   | O(n) | O(n)  | ❌ Avoid  |

---

# 🎯 Best Choice

👉 Use:

```python
sum(map(int, s))
```

or simple loop

👉 *Just iterate through digits and add them.*
