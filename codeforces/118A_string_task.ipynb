```python
s = "Codeforces"
s = s.lower()
# print(s)

v = "aoyeoi"

lst = []

for i in range(len(s)):
        if s[i] not in v:
            lst.append(s[i])
print(lst)  


x = "".join(lst)
print(x)
sum = ""
for j in range(len(x)):
      sum += "."+ x[j]

print(sum)
```

```python
s = input().lower()

v = "aoyeui"

result = ""

for ch in s:
    if ch not in v:
        result += "." + ch

print(result)
```
