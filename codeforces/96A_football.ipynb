```python
s = "1000000001"

print(s.count("1"))
print(s.count("0"))

for i in range(len(s)):
    count = 0
    if s[i] == "0":
        count += 1
print(count)            
```

```python
s = "1000000001"


if "0000000" in s:
    print("YES")
else:
    print("NO")    

if "11111111" in s:
    print("YES")
else:
    print("NO")    
```

```python
s = "1000000001"

if "0000000" in s or "1111111" in s:
    print("YES")
else:
    print("NO")    
```
