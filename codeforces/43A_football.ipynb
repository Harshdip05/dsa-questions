```python
n = int(input())
teams = [input() for _ in range(n)]

unique = list(set(teams))

# Since there are only 2 teams
if len(unique) == 1:
    print(unique[0])
else:
    if teams.count(unique[0]) > teams.count(unique[1]):
        print(unique[0])
    else:
        print(unique[1])
```

```python
from collections import Counter

n = int(input())
teams = [input() for _ in range(n)]

count = Counter(teams)

# get team with maximum goals
print(count.most_common(1)[0][0])
```
