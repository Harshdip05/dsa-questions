```python
A = [20, 1, 9, 4]

for i in range(len(A)):
    for j in range(i+1,len(A)):
        print(A[i]+A[j]+j-i)
```

```python
A = [20, 1, 9, 4]

mini = float("inf")
for i in range(len(A)):
    for j in range(i+1,len(A)):
        d = (A[i]+A[j]+j-i)
        mini = min(d,mini)

print(mini)
```

```python
T = int(input())

for _ in range(T):
    
    N = int(input())
    A = list(map(int, input().split()))
    
    mini = float("inf")
    for i in range(N):
        for j in range(i + 1, N):
            d = A[i] + A[j] + j - i
            mini = min(mini, d)
    
    print(mini)   # ✅ inside loop
```
