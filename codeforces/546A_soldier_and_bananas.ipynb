| Method  | Time | Space  |
| ------- | ---- | ------ |
| Formula | O(1) | O(1) ✅ |
| Loop    | O(w) | O(1)   |
| sum()   | O(w) | O(1)   |


```python
k = 3 
n = 17 
w = 4

sum = 0
for i in range(1,w+1):
    sum += (i*k)
    # print(sum)

print(sum)    

if n<sum:
    print(sum-n)
else:
    print(0)


```

```python
k, n, w = map(int, input().split())

total_cost = k * (w * (w + 1) // 2)

print(max(0, total_cost - n))
```

```python
k, n, w = map(int, input().split())

total_cost = sum([i * k for i in range(1, w + 1)])

print(max(0, total_cost - n))
```
