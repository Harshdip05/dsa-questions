```python
## F. REVERSING

N = int(input())

arr = list(map(int,input().split()))

left = 0
right = N-1

while(left<right):
    arr[left],arr[right] = arr[right],arr[left]
    
    left +=1
    right -=1
    
print(arr) 
print(*arr)   
```

```python
N = int(input())
arr = list(map(int, input().split()))

arr.reverse()

print(*arr)
```

*arr means

The * is called the unpacking operator.

It takes the list arr and unpacks its elements as separate arguments to print().

### E - Lowest Number

```python
N = int(input())
print(N)
A = list(map(int,input().split()))
print(A)

min_num = float('inf')
index = 0

for i in range(len(A)):
    if min_num > A[i]:
        min_num = A[i]
        index = i
        # break
  
print(min_num,index+1)  
```

### F - Palindrome Array

```python
N = int(input())
A = list(map(int,input().split()))

if A==A[::-1]:
    print("YES")
else:
    print("NO")    
```
