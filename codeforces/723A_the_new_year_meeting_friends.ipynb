# 🔹 Problem Recap

Given:

```python
x1, x2, x3
```

👉 Find minimum total distance to meet at one point

---

# 🔥 Method 1: Using Median (Best & Standard)

```python
x = sorted(map(int, input().split()))
median = x[1]

result = abs(x[0] - median) + abs(x[1] - median) + abs(x[2] - median)

print(result)
```

### ✅ Why:

* Median minimizes total distance

---

# 🔥 Method 2: Direct Formula (Shortest Code)

```python
x = sorted(map(int, input().split()))
print(x[2] - x[0])
```

### ✅ Why:

👉 For 3 points:

```
distance = max - min
```

---

# 🔥 Method 3: Try All Possible Points (Brute Force)

```python
x1, x2, x3 = map(int, input().split())

ans = float('inf')

for p in range(1, 101):   # possible positions
    dist = abs(x1 - p) + abs(x2 - p) + abs(x3 - p)
    ans = min(ans, dist)

print(ans)
```

### ✅ Why:

* Check every possible meeting point
* Pick minimum

### ❌ Slower but good for understanding

---

# 🔥 Method 4: Without Sorting (Using Math)

```python
x1, x2, x3 = map(int, input().split())

maxi = max(x1, x2, x3)
mini = min(x1, x2, x3)

print(maxi - mini)
```

---

# 🔥 Method 5: Manual Median (No Sorting)

```python
x1, x2, x3 = map(int, input().split())

if (x1 <= x2 <= x3) or (x3 <= x2 <= x1):
    median = x2
elif (x2 <= x1 <= x3) or (x3 <= x1 <= x2):
    median = x1
else:
    median = x3

result = abs(x1 - median) + abs(x2 - median) + abs(x3 - median)

print(result)
```

---

# 🔹 Comparison

| Method        | Idea         | Complexity | Best Use   |
| ------------- | ------------ | ---------- | ---------- |
| Median        | middle value | O(1)       | ✅ Best     |
| max-min       | shortcut     | O(1)       | ✅ Shortest |
| Brute force   | try all      | O(n)       | learning   |
| Manual median | logic        | O(1)       | interviews |

---

# 🧠 Key Insight

> All methods give same result because:
> 👉 Optimal meeting point = **median**
