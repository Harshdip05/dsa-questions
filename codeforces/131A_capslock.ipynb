```python
words = input()

def manual_swap(s):
    result = ""
    for ch in s:
        if ch.islower():
            result += ch.upper()
        else:
            result += ch.lower()
    return result

if words.isupper():
    print(manual_swap(words))
elif len(words) == 1 and words[0].islower():
    print(manual_swap(words))
elif words[0].islower() and words[1:].isupper():
    print(manual_swap(words))
else:
    print(words)
```
