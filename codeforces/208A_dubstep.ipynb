```python
# 

s = input().strip()

# Replace WUB with space
s = s.replace("WUB", " ")

# Split and rejoin to remove extra spaces
result = " ".join(s.split())

print(result)
```

```python
s = input().strip()

words = s.split("WUB")

# Remove empty strings
words = [w for w in words if w]

print(" ".join(words))
```

```python
s = input().strip()
i = 0
result = []

while i < len(s):
    if s[i:i+3] == "WUB":
        result.append(" ")
        i += 3
    else:
        result.append(s[i])
        i += 1

# Clean extra spaces
print(" ".join("".join(result).split()))
```
