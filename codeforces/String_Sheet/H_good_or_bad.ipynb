This is a classic **substring detection** problem:
👉 Check if `"010"` **or** `"101"` exists in the string.

Let’s solve it in **multiple ways**, from simplest to more advanced, with **time & space analysis**.

---

# ✅ ✅ Method 1: Built-in substring check (Best & simplest)

```python
def solve():
    t = int(input())
    for _ in range(t):
        s = input().strip()
        if "010" in s or "101" in s:
            print("Good")
        else:
            print("Bad")
```

### ⏱️ Complexity

* Time: **O(n)** per test case
* Space: **O(1)**

### 💡 Why this is best

* Python uses optimized substring search (very fast in practice)
* Clean and readable

---

# ✅ Method 2: Sliding window (manual check)

Instead of slicing, just check every length-3 window.

```python
def solve():
    t = int(input())
    for _ in range(t):
        s = input().strip()
        found = False
        
        for i in range(len(s) - 2):
            if s[i:i+3] == "010" or s[i:i+3] == "101":
                found = True
                break
        
        print("Good" if found else "Bad")
```

### ⏱️ Complexity

* Time: **O(n)**
* Space: **O(1)**

---

# ✅ Method 3: No slicing (most efficient low-level way)

Avoid creating substrings entirely:

```python
def solve():
    t = int(input())
    for _ in range(t):
        s = input().strip()
        found = False
        
        for i in range(len(s) - 2):
            if (s[i] == '0' and s[i+1] == '1' and s[i+2] == '0') or \
               (s[i] == '1' and s[i+1] == '0' and s[i+2] == '1'):
                found = True
                break
        
        print("Good" if found else "Bad")
```

### ⏱️ Complexity

* Time: **O(n)**
* Space: **O(1)**
* 🚀 Slightly faster than slicing version

---

# ❌ Method 4: Brute force (DON’T use)

Generate all substrings:

```python
for i in range(len(s)):
    for j in range(i, len(s)):
        if s[i:j+1] == "010" or s[i:j+1] == "101":
            ...
```

### ⏱️ Complexity

* Time: **O(n³)** ❌ (too slow for n = 100000)
* Space: **O(n)** (due to slicing)

👉 This will **TLE (Time Limit Exceeded)**

---

# 🧠 Key Observation (Important Trick)

You only need to check substrings of length **3**.

👉 Not all substrings — just windows of size 3.

---

# 🔥 Optimal Thinking

Instead of:

> “Check all substrings”

Think:

> “Check every group of 3 consecutive characters”
