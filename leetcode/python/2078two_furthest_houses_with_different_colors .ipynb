This is a nice problem because there are **multiple ways to solve it**, from brute force to optimized. Let’s go step by step and compare them.

---

# 🧠 Problem Summary

We need to find:

👉 Maximum distance ( |i - j| ) such that
👉 `colors[i] != colors[j]`

---

# ✅ Method 1: Brute Force (Check All Pairs)

### 💡 Idea

* Try every pair ((i, j))
* If colors differ → update max distance

### 💻 Code

```python
def maxDistance(colors):
    n = len(colors)
    max_dist = 0
    
    for i in range(n):
        for j in range(n):
            if colors[i] != colors[j]:
                max_dist = max(max_dist, abs(i - j))
    
    return max_dist
```

### ⏱️ Complexity

* Time: **O(n²)**
* Space: **O(1)**

👉 Works fine for small n, but inefficient

---

# 🚀 Method 2: Optimized (Compare Ends)

### 💡 Key Insight

Maximum distance will **always involve either**:

* First house (index 0), or
* Last house (index n-1)

So:

* Compare first house with all others
* Compare last house with all others

---

### 💻 Code

```python
def maxDistance(colors):
    n = len(colors)
    max_dist = 0
    
    # Compare with first house
    for j in range(n):
        if colors[j] != colors[0]:
            max_dist = max(max_dist, j)
    
    # Compare with last house
    for i in range(n):
        if colors[i] != colors[n-1]:
            max_dist = max(max_dist, n - 1 - i)
    
    return max_dist
```

---

### 🔍 Example

Input: `[1,1,1,6,1,1,1]`

* Compare with first (1):

  * index 3 → different → distance = 3
* Compare with last (1):

  * index 3 → different → distance = 3

👉 Answer = **3**

---

### ⏱️ Complexity

* Time: **O(n)**
* Space: **O(1)**

👉 Much better than brute force

---

# ⚡ Method 3: Two-Pointer Style (Cleaner)

### 💡 Idea

* Start from left and right ends
* Find first mismatch from each side

---

### 💻 Code

```python
def maxDistance(colors):
    n = len(colors)
    
    # from left
    left = 0
    while colors[left] == colors[n - 1]:
        left += 1
    
    # from right
    right = n - 1
    while colors[right] == colors[0]:
        right -= 1
    
    return max(n - 1 - left, right)
```

---

### ⏱️ Complexity

* Time: **O(n)**
* Space: **O(1)**

👉 Same efficiency, slightly cleaner

---

# 📊 Comparison

| Method      | Time Complexity | Space | Notes           |
| ----------- | --------------- | ----- | --------------- |
| Brute Force | O(n²)           | O(1)  | Simple but slow |
| Optimized   | O(n)            | O(1)  | Best approach   |
| Two-pointer | O(n)            | O(1)  | Cleaner version |

---

# 🎯 Key Insight (Important)

The **maximum distance must involve one of the ends**:

* Either `(0, j)`
* Or `(i, n-1)`

👉 That’s why we don’t need to check all pairs
