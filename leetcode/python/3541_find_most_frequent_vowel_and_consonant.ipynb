```python
from collections import Counter

s = "successes"
vowels = "aeiou"

consonants = Counter()
print(consonants)

for ch in s:
    if ch not in vowels:
        consonants[ch] += 1

print(consonants)




vowels_count = Counter()

for ch in s:
    if ch in vowels:
        vowels_count[ch] += 1

print(vowels_count)



print(max(consonants.values()))
print(max(vowels_count.values()))
```
