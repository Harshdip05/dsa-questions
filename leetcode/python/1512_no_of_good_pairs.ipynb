````python
from collections import Counter
from math import comb


class Solution:
    def numIdenticalPairs(self, arr: list[int]) -> int:
        freq_counter = Counter(Counter(arr).values())
        return sum(comb(freq, 2) * times for freq, times in freq_counter.items())
````

# 🔹 Problem (in simple words)

👉 Count pairs `(i, j)` such that:

* `i < j`
* `nums[i] == nums[j]`

---

# 🔹 Example

```python
nums = [1,2,3,1,1,3]
```

✅ Output:

```text
4
```

---

# 🔴 Method 1: Brute Force (double loop)

## 🔹 Idea

Check all pairs

---

## 🔹 Code

```python
def good_pairs(nums):
    count = 0
    n = len(nums)
    
    for i in range(n):
        for j in range(i+1, n):
            if nums[i] == nums[j]:
                count += 1
    
    return count

print(good_pairs([1,2,3,1,1,3]))
```

---

## ✅ Output

```text
4
```

---

## ⏱ Complexity

* Time: **O(n²)** ❌ slow
* Space: **O(1)** ✅

---

# 🟡 Method 2: HashMap (Best solution)

## 🔹 Idea

Count frequency while iterating

👉 If number already seen `k` times → it forms `k` new pairs

---

## 🔹 Code

```python
def good_pairs(nums):
    freq = {}
    count = 0
    
    for num in nums:
        if num in freq:
            count += freq[num]
            freq[num] += 1
        else:
            freq[num] = 1
    
    return count

print(good_pairs([1,2,3,1,1,3]))
```

---

## ✅ Output

```text
4
```

---

## 🔹 Step intuition

For `[1,2,3,1,1,3]`:

| Num | Seen before | New pairs |
| --- | ----------- | --------- |
| 1   | 0           | 0         |
| 2   | 0           | 0         |
| 3   | 0           | 0         |
| 1   | 1           | +1        |
| 1   | 2           | +2        |
| 3   | 1           | +1        |

Total = **4**

---

## ⏱ Complexity

* Time: **O(n)** ✅ fastest
* Space: **O(n)**

---

# 🟢 Method 3: Using combination formula

## 🔹 Idea

If a number appears `f` times:

👉 pairs =
[
f(f-1)/2
]

---

## 🔹 Code

```python
def good_pairs(nums):
    from collections import Counter
    
    freq = Counter(nums)
    count = 0
    
    for f in freq.values():
        count += f * (f - 1) // 2
    
    return count

print(good_pairs([1,2,3,1,1,3]))
```

---

## ✅ Output

```text
4
```

---

## ⏱ Complexity

* Time: **O(n)**
* Space: **O(n)**

---

# 🔵 Method 4: Using array (since values ≤ 100)

## 🔹 Idea

Use fixed-size list instead of dictionary

## 🔹 Code

```python
def good_pairs(nums):

print(good_pairs([1,2,3,1,1,3]))
```

---

## ⏱ Complexity

* Time: **O(n)**
* Space: **O(1)** ✅ (constant size 101)

```python
freq = Counter(nums)
```

👉 This counts how many times each number appears.

For:

```python
nums = [1,2,3,1,1,3]
```

👉 `freq` becomes:



# 🔹 Step 3: Formula

```python
count += f * (f - 1) // 2
```

👉 This calculates:

> How many pairs can be made from `f` same numbers

---

## 🔍 Why this formula?

If a number appears `f` times:

👉 number of pairs =
[
\frac{f(f-1)}{2}
]

---

# 🔹 Step-by-step calculation

### For number `1` → appears 3 times

```text
f = 3
pairs = 3 * 2 / 2 = 3
```

Pairs:

```text
(1,1), (1,1), (1,1)
```

---

### For number `2` → appears 1 time

```text
f = 1
pairs = 1 * 0 / 2 = 0
```

---

### For number `3` → appears 2 times

```text
f = 2
pairs = 2 * 1 / 2 = 1
```

---

# 🔹 Total

```text
3 + 0 + 1 = 4
```

👉 If a number appears multiple times:
