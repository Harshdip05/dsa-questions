# 1) Manual sort using Bubble Sort

```python
def targetIndices(nums, target):
    n = len(nums)

    # bubble sort
    for i in range(n):
        for j in range(n - i - 1):
            if nums[j] > nums[j + 1]:
                nums[j], nums[j + 1] = nums[j + 1], nums[j]

    result = []
    for i in range(n):
        if nums[i] == target:
            result.append(i)

    return result
```

**Time Complexity:**

* Bubble sort: `O(n^2)`
* Scan for target: `O(n)`
* Total: `O(n^2)`

**Space Complexity:**

* `O(1)` extra space
* excluding output array

# 4) Sort + Binary Search

This is what you asked for earlier.

```python
def targetIndices(nums, target):
    nums.sort()

    def first_occurrence(nums, target):
        left, right = 0, len(nums) - 1
        ans = -1

        while left <= right:
            mid = (left + right) // 2

            if nums[mid] == target:
                ans = mid
                right = mid - 1
            elif nums[mid] < target:
                left = mid + 1
            else:
                right = mid - 1

        return ans

    def last_occurrence(nums, target):
        left, right = 0, len(nums) - 1
        ans = -1

        while left <= right:
            mid = (left + right) // 2

            if nums[mid] == target:
                ans = mid
                left = mid + 1
            elif nums[mid] < target:
                left = mid + 1
            else:
                right = mid - 1

        return ans

    first = first_occurrence(nums, target)
    if first == -1:
        return []

    last = last_occurrence(nums, target)
    return list(range(first, last + 1))
```

**Time Complexity:**

* Sort: `O(n log n)`
* First binary search: `O(log n)`
* Last binary search: `O(log n)`
* Build answer: `Ok`, where `k` = count of target
* Total: `O(n log n)`

**Space Complexity:**

* `O(1)` extra space
* excluding output array

# 5) Sort + `bisect`

```python
from bisect import bisect_left, bisect_right

def targetIndices(nums, target):
    nums.sort()
```

# 6) Optimal counting method, no sorting

This is the best method.

```python
def targetIndices(nums, target):
    less = 0
    equal = 0

    for num in nums:
        if num < target:
            less += 1
        elif num == target:
            equal += 1

    result = []
    for i in range(less, less + equal):
        result.append(i)

    return result
```

**Why it works:**

* `less` = how many numbers are smaller than target
* `equal` = how many numbers equal target
* after sorting, target will start from index `less` and end at `less + equal - 1`

**Time Complexity:**

* Count pass: `O(n)`
* Build result: `O(k)`
* Total: `O(n)`

**Space Complexity:**

* `O(1)` extra space
* excluding output array

# 7) Frequency array / counting sort idea

Since constraints are small: `1 <= nums[i], target <= 100`

```python
def targetIndices(nums, target):
    freq = [0] * 101

    for num in nums:
        freq[num] += 1

    less = 0
    for i in range(target):
        less += freq[i]

    equal = freq[target]

    result = []
    for i in range(less, less + equal):
        result.append(i)

    return result
```

**Time Complexity:**

* Build frequency: `O(n)`
* Count less than target: `O(100)`
* Build result: `O(k)`
* Total: `O(n + 100)` → effectively `O(n)`

**Space Complexity:**

* `O(100)` → effectively `O(1)`
