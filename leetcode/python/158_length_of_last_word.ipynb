Given a string s consisting of words and spaces, return the length of the last word in the string.

A word is a maximal substring consisting of non-space characters only.

 

Example 1:

Input: s = "Hello World"
Output: 5
Explanation: The last word is "World" with length 5.
Example 2:

Input: s = "   fly me   to   the moon  "
Output: 4
Explanation: The last word is "moon" with length 4.
Example 3:

Input: s = "luffy is still joyboy"
Output: 6
Explanation: The last word is "joyboy" with length 6.
 

Constraints:

1 <= s.length <= 104
s consists of only English letters and spaces ' '.
There will be at least one word in s.

Great question — this is a very common confusion 👍
Let’s clearly see the difference between `split()` and `split(" ")`.

---

## 🔹 1. `split()`

* Splits on **any whitespace**
* Treats **multiple spaces as one**
* Also handles **tabs (`\t`) and newlines (`\n`)**

### Example:

```python
text = "Hello   world\tPython\ncode"
print(text.split())
```

Output:

```id="out1"
['Hello', 'world', 'Python', 'code']
```

👉 It automatically ignores extra spaces and different whitespace types.

---

## 🔹 2. `split(" ")`

* Splits **only on a single space character `' '`**
* Does **NOT ignore multiple spaces**
* Can produce **empty strings**

### Example:

```python
text = "Hello   world"
print(text.split(" "))
```

Output:

```id="out2"
['Hello', '', '', 'world']
```

👉 Each space is treated separately, so extra spaces create empty entries.

---

## 🔥 Key Differences

| Feature                 | `split()`              | `split(" ")`                |
| ----------------------- | ---------------------- | --------------------------- |
| Handles multiple spaces | ✅ Yes (ignores extras) | ❌ No                        |
| Handles tabs/newlines   | ✅ Yes                  | ❌ No                        |
| Empty strings in result | ❌ No                   | ✅ Yes                       |
| Common use              | General word splitting | Exact space-based splitting |

---

## 🧠 Simple Rule:

* Use **`split()`** → when working with normal text (recommended)
* Use **`split(" ")`** → when you specifically care about exact spaces

````python
class Solution:
    def lengthOfLastWord(self, s: str) -> int:
        # s = 
        x = s.split()
        x = x[-1]
        y = len(x)
        return y
````
