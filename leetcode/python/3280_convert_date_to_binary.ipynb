```python
date = "2080-02-29"

x = date.split("-")
print(x)

x1 = x[0]
z1 = bin(int(x1))[2::]
print(z1)
print(bin(int(x1))[2::])

x2 = x[1]
print(bin(int(x2))[2::])

x3 = x[2]
print(bin(int(x3))[2::])

print(str(bin(int(x1))[2::])+"-"+str(bin(int(x2))[2::])+"-"+str(bin(int(x3))[2::]))


```

```python
date = "2080-02-29"

x = date.split("-")
print(x)

sum = ""
for i in range(len(x)):
    x1 = str(bin(int(x[i]))[2::])
    # print(x1)
    sum += x1 + "-"
print(sum[:-1])
```
