## ✅ 1. Optimal Math (if-else)

```python id="s1a2b3"
class Solution:
    def smallestEvenMultiple(self, n: int) -> int:
        return n if n % 2 == 0 else 2 * n
```

**Idea:**

* Even → already valid
* Odd → multiply by 2

**Complexity:**

* Time: `O(1)`
* Space: `O(1)`

---

## ✅ 2. Using `math.lcm` (Best Pythonic way)

```python id="s4d5e6"
import math

class Solution:
    def smallestEvenMultiple(self, n: int) -> int:
        return math.lcm(2, n)
```

**Idea:**
Directly compute **LCM of 2 and n**

**Complexity:**

* Time: `O(log n)`
* Space: `O(1)`

👉 This is the **cleanest one-liner**

---

## ✅ 3. Using LCM Formula (Manual)

```python id="s7g8h9"
import math

class Solution:
    def smallestEvenMultiple(self, n: int) -> int:
        return (2 * n) // math.gcd(2, n)
```

**Idea:**
[
LCM = \frac{a \times b}{GCD(a, b)}
]

**Complexity:**

* Time: `O(log n)`
* Space: `O(1)`

---

## ✅ 4. While Loop (your approach, improved)

```python id="s10i11"
class Solution:
    def smallestEvenMultiple(self, n: int) -> int:
        i = n
        while True:
            if i % 2 == 0:
                return i
            i += n
```

**Idea:**
Check multiples of `n` until one is even

**Complexity:**

* Time: `O(1)` (max 2 iterations)
* Space: `O(1)`

---

## ✅ 5. Brute Force

```python id="s12j13"
class Solution:
    def smallestEvenMultiple(self, n: int) -> int:
        for i in range(1, 2 * n + 1):
            if i % 2 == 0 and i % n == 0:
                return i
```

**Idea:**
Try all numbers until condition matches

**Complexity:**

* Time: `O(n)`
* Space: `O(1)`
