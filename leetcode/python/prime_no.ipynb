```python
# 0 and negative numbers are NOT prime.
```

```python
# A number is prime if:
        # It is greater than 1
        # It has only 2 factors → 1 and itself
```

```python
n = int(input("Enter number: "))

if n <= 1:
    print("Not Prime")
else:
    for i in range(2, n):
        if n % i == 0:
            print("Not Prime")
            break
    else:                       ## This else runs only if loop finishes without break
        print("Prime")
```

```python
def is_prime(n):
    if n <= 1:
        return "Not Prime"
    
    for i in range(2, n):
        if n % i == 0:
            return "Not Prime"
    
    return "Prime"


# taking input
num = int(input("Enter number: "))
print(is_prime(num))
```
