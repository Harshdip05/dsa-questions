| Method      | Time | Space | Notes          |
| ----------- | ---- | ----- | -------------- |
| Counter     | O(n) | O(1)  | Cleanest ⭐     |
| Dictionary  | O(n) | O(1)  | Most common    |
| Manual vars | O(n) | O(1)  | Fastest        |
| `count()`   | O(n) | O(1)  | Less efficient |


```python
from collections import Counter

def max_balloons(text):
    count = Counter(text)
    return min(count["b"],count["a"],count["l"]/2,count["o"]/2,count["l"])

text = input()
result = max_balloons(text)
print(result)
```

```python
def max_balloons(text):
    freq = {}

    # count frequency of each character
    for ch in text:
        if ch in freq:
            freq[ch] += 1
        else:
            freq[ch] = 1

    # calculate result
    return min(
        freq.get('b', 0),
        freq.get('a', 0),
        freq.get('l', 0) // 2,
        freq.get('o', 0) // 2,
        freq.get('n', 0)
    )

text = input()
print(max_balloons(text))
```
