```python
# Using Sliding Window (for 'L')

def checkRecord(s):
    if s.count('A') >= 2:
        return False

    for i in range(len(s) - 2):
        if s[i] == s[i+1] == s[i+2] == 'L':
            return False

    return True
```

```python
# using regular exp

import re

def checkRecord(s):
    return not (re.search("A.*A", s) or re.search("LLL", s))
```

```python
# using loop

s ="ALLAPPL"

absent = 0
late = 0


for  i in range(len(s)):

    if  s[i] == "A": 
        absent +=1
        if absent>=2:
          print("False")
          break

    if s[i]=="L":
        late += 1
        if late>=3:
            print("False")
            break
    else: 
        late = 0

else:
    print("True")
```

```python
s = "ALLAPPL"

absent = 0
late = 0

for ch in s:
    if ch == "A":
        absent += 1
        if absent >= 2:
            print("False")
            break

    if ch == "L":
        late += 1
        if late >= 3:
            print("False")
            break
    else:
        late = 0
else:
    print("True")
```
