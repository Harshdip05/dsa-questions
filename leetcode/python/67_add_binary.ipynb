Given two binary strings a and b, return their sum as a binary string.


Example 1:

Input: a = "11", b = "1"
Output: "100"
Example 2:

Input: a = "1010", b = "1011"
Output: "10101"

```python
int("10")   # → 10 (string → decimal)
```

```python
int("11", 2)   # binary → decimal
```

```python
int("101", 2)   # → 5
int("A", 16)    # → 10 (hexadecimal)
int("10", 8)    # → 8 (octal)
```

```python
print(bin(5)[2:])
bin(5)
```

```python
format(5, 'b')
```

| Function        | Purpose                  | Example                |
| --------------- | ------------------------ | ---------------------- |
| `int(x)`        | string → decimal         | `int("10") → 10`       |
| `int(x, 2)`     | binary → decimal         | `int("11",2) → 3`      |
| `bin(x)`        | decimal → binary         | `bin(3) → '0b11'`      |
| `format(x,'b')` | decimal → binary (clean) | `format(3,'b') → '11'` |

bin() → adds "0b"

format(n, 'b') → no "0b"

```python
a = "11"
b = "1"

result = bin(int(a, 2) + int(b, 2))
print(result)
print(result[2::])


result2 = bin(int(a, 2)) + bin(int(b, 2))   ## This is string concatenation, NOT binary addition ❌
print(result2)
print(result2[2::])

xx = int(result, 2)
print(xx)
```
