We define the usage of capitals in a word to be right when one of the following cases holds:

All letters in this word are capitals, like "USA".
All letters in this word are not capitals, like "leetcode".
Only the first letter in this word is capital, like "Google".
Given a string word, return true if the usage of capitals in it is right.

 

Example 1:

Input: word = "USA"
Output: true
Example 2:

Input: word = "FlaG"
Output: false
 

Constraints:

1 <= word.length <= 100
word consists of lowercase and uppercase English letters.

```python
# Methods like .isupper(), .islower(), .istitle() must be called with ()
# Without (), Python sees the method object, which is always “truthy”
```

```python
word = "FlaG"

if word.isupper() or word.islower() or word.istitle():
    print(True)
else:
    print(False)
```

```python
word = "USA"

if word.upper() or word.lower or word.title:
    print(True)
else:
    print(False)
```

```python
# A mixed-case word is not all uppercase, not all lowercase, and not title case.``

word = "FlaG"

if not (word.isupper() or word.islower() or word.istitle()):
    print("Mixed case")
else:
    print("Not mixed case")
```

```python
class Solution:
    def detectCapitalUse(self, word: str) -> bool:
        n = len(word)
        count = 0
        for ch in word:
            if ch.isupper():
                count += 1

        if count == n:
            return True
        if count == 0:
            return True
        if word[0].isupper() and count == 1:
            return True
        return False
```
