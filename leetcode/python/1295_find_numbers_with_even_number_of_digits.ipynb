### Given an array nums of integers, return how many of them contain an even number of digits.

```python
nums = [12,34,11111,222,5]

# print(len(str(nums[1])))

count = 0

for i in range(len(nums)):
    if len(str(nums[i])) % 2 == 0 :
        count += 1

print(count)        
```

```python
class Solution:
    def findNumbers(self, nums: List[int]) -> int:
     
        digit=0

        for i in nums:
            count=0
            while i>0:
               i=i//10
               count+=1

            if count%2==0:
                digit+=1
        return digit
```

Without String (Pure Division)
🔹 Idea

Keep dividing by 10 until number becomes 0


```python
def findNumbers(nums):
    count = 0
    for x in nums:
        digits = 0
        n = x
        while n > 0:
            n //= 10
            digits += 1
        if digits % 2 == 0:
            count += 1
    return count
```

Using Range Check (Fastest 🚀)
🔹 Idea

Digits are even if number is:

2 digits → 10–99
4 digits → 1000–9999
6 digits → 100000–999999

```python
def findNumbers(nums):
    count = 0
    for x in nums:
        if 10 <= x <= 99 or 1000 <= x <= 9999 or x == 100000:
            count += 1
    return count
```

```python
import math

def findNumbers(nums):
    count = 0
    for x in nums:
        digits = int(math.log10(x)) + 1
        if digits % 2 == 0:
            count += 1
    return count
```
