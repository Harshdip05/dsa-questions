```python
nums = [12,21,45,33,54]

lst = []
for x in nums:
    lst.append(str(x))

print(lst)

seen = {}
new = []

for i in range(len(lst)):
    rev = int(lst[i][::-1])

    if rev in seen:
        new.append(i - seen[rev])

    seen[int(lst[i])] = i   # store original number

print(new)

print(min(new) if new else -1)
```
