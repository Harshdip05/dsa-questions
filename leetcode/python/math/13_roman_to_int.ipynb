# 🔥 1. Validate Roman Numeral

## 🧩 Problem

Given a string `s`, return **True if it is a valid Roman numeral**, else False.

### Examples:

```text
"III"   → True
"IV"    → True
"IC"    → False
"VV"    → False
```

---

## 🧠 Key Rules

1. Valid subtraction:

   * I → V, X
   * X → L, C
   * C → D, M

2. Repetition:

   * I, X, C, M → max 3 times
   * V, L, D → cannot repeat

---

## 🚀 Code

```python
def isValidRoman(s):
    valid_sub = {
        'I': ['V', 'X'],
        'X': ['L', 'C'],
        'C': ['D', 'M']
    }

    repeat_count = 1

    for i in range(len(s)):
        # repetition check
        if i > 0 and s[i] == s[i-1]:
            repeat_count += 1

            if s[i] in ['V', 'L', 'D'] or repeat_count > 3:
                return False
        else:
            repeat_count = 1

        # subtraction check
        if i < len(s)-1:
            if s[i] < s[i+1]:
                if s[i] not in valid_sub or s[i+1] not in valid_sub[s[i]]:
                    return False

    return True
```

# 🔥 4. Generate Roman Numbers up to N

## 🧩 Problem

Given integer `n`, return all Roman numerals from `1 → n`

---

## 🚀 Code

```python
def intToRoman(num):
    values = [
        (1000, 'M'), (900, 'CM'),
        (500, 'D'), (400, 'CD'),
        (100, 'C'), (90, 'XC'),
        (50, 'L'), (40, 'XL'),
        (10, 'X'), (9, 'IX'),
        (5, 'V'), (4, 'IV'),
        (1, 'I')
    ]

    result = ""

    for value, symbol in values:
        while num >= value:
            result += symbol
            num -= value

    return result


def generateRoman(n):
    res = []
    for i in range(1, n+1):
        res.append(intToRoman(i))
    return res
```
