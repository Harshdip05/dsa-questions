| ---------- | -------------------- | ------------------ |
| Dictionary | lookup (Roman → int) | ordered processing |
| List/Tuple | greedy construction  | —                  |


```python
def intToRoman(num):
    pairs = [
        (1000, "M"), (900, "CM"),
        (500, "D"), (400, "CD"),
        (100, "C"), (90, "XC"),
        (50, "L"), (40, "XL"),
        (10, "X"), (9, "IX"),
        (5, "V"), (4, "IV"),
        (1, "I")
    ]

    result = ""

    for value, symbol in pairs:
        while num >= value:
            result += symbol
            num -= value

    return result
```

```python
def intToRoman(num):
    values = [1000, 900, 500, 400, 100, 90, 50, 40,
              10, 9, 5, 4, 1]

    symbols = ["M", "CM", "D", "CD", "C", "XC", "L", "XL",
               "X", "IX", "V", "IV", "I"]

    result = ""

    for i in range(len(values)):
        while num >= values[i]:
            result += symbols[i]
            num -= values[i]

    return result
```

```python
def intToRoman(num):
    roman_map = {
        1000: "M", 900: "CM",
        500: "D", 400: "CD",
        100: "C", 90: "XC",
        50: "L", 40: "XL",
        10: "X", 9: "IX",
        5: "V", 4: "IV",
        1: "I"
    }

    result = ""

    # sort keys in descending order
    for value in sorted(roman_map.keys(), reverse=True):
        while num >= value:
            result += roman_map[value]
            num -= value

    return result
```

```python
def intToRoman(num):
    thousands = ["", "M", "MM", "MMM"]
    hundreds = ["", "C", "CC", "CCC", "CD", "D", "DC", "DCC", "DCCC", "CM"]
    tens = ["", "X", "XX", "XXX", "XL", "L", "LX", "LXX", "LXXX", "XC"]
    ones = ["", "I", "II", "III", "IV", "V", "VI", "VII", "VIII", "IX"]

    return (
        thousands[num // 1000] +
        hundreds[(num % 1000) // 100] +
        tens[(num % 100) // 10] +
        ones[num % 10]
    )
```
