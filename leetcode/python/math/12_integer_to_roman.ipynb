````python
def intToRoman(num):
    roman_map = {
        1000: "M", 900: "CM",
        500: "D", 400: "CD",
        100: "C", 90: "XC",
        50: "L", 40: "XL",
        10: "X", 9: "IX",
        5: "V", 4: "IV",
        1: "I"
    }

    result = ""

    # sort keys in descending order
    for value in sorted(roman_map.keys(), reverse=True):
        while num >= value:
            result += roman_map[value]
            num -= value

    return result
````

````python
def intToRoman(num):
    thousands = ["", "M", "MM", "MMM"]
    hundreds = ["", "C", "CC", "CCC", "CD", "D", "DC", "DCC", "DCCC", "CM"]
    tens = ["", "X", "XX", "XXX", "XL", "L", "LX", "LXX", "LXXX", "XC"]
    ones = ["", "I", "II", "III", "IV", "V", "VI", "VII", "VIII", "IX"]

    return (
        thousands[num // 1000] +
        hundreds[(num % 1000) // 100] +
        tens[(num % 100) // 10] +
        ones[num % 10]
    )
````

👉 **Short answer: NO, it is NOT O(n²)**
Let’s break it properly.

---

# 🧠 Key Idea

👉 Time complexity depends on **how many total operations happen**, not just nested loops.

---

# 🔍 Look at the code again

```python
for value in values:          # runs 13 times (fixed)
    while num >= value:       # runs multiple times
        num -= value
```

---

# 🔥 Important Observation

### Outer loop:

* Runs **13 times only** (fixed list size)
  👉 constant

---

### Inner loop:

* Each time it runs → **num decreases**

👉 This is the key point

---

# 🧠 Think Like This

Let’s say:

```python
num = 3999
```

Every iteration:

* num is reduced (1000 → 900 → 100 → ...)

👉 Eventually num becomes 0

---

## 💡 Total number of `while` executions

Worst case:

* 3000 → MMM → 3 times
* 300 → CCC → 3 times
* 30 → XXX → 3 times
* 3 → III → 3 times

👉 Total ≈ **12–15 operations**

---

# 🚫 Why it is NOT O(n²)

For O(n²):

* Inner loop must run **n times for each outer loop**

Example:

```python
for i in range(n):
    for j in range(n):
```

👉 That gives n × n

---

# ✅ Here what happens?

* Outer loop = constant (13)
* Inner loop = total work proportional to **num decreasing**

👉 So total work is **linear in num (but num is bounded)**

---

# ⚡ Final Complexity

## General case:

👉 **O(num)**

## Given constraint (1 ≤ num ≤ 3999):

👉 **O(1)**

---

# 🔥 Best Way to Explain in Interview

Say:

> “Although there is a nested loop, the inner loop reduces the input value each time, so the total number of iterations is bounded. Since the number range is limited (≤ 3999), the overall complexity is constant time.”

* ₹100 → 9 times
* ₹10 → 9 times

👉 Total steps = small number
👉 Not n²

# 🚀 Key Takeaway

👉 Nested loops ≠ always O(n²)
👉 You must check:

* Does inner loop depend on outer loop?
* Or is it shrinking input?
