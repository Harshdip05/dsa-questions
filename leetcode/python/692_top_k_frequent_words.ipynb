```python
from itertools import permutations

s = "abccccdd"

for x in permutations(s):
    z = "".join(x)
    # print(type(z))
    if z == z[::-1]:
        print(z)
```

```python
from collections import Counter

s = "abccccdd"
count = Counter(s)

odd = sum(1 for v in count.values() if v % 2 != 0)

if odd > 1:
    print("No palindrome possible")
else:

    print("Palindrome possible")
```

```python
from collections import Counter

words = ["i","love","leetcode","i","love","coding"]
k = 2

c = Counter(words)
print(c)
print(type(c))

x = (c.most_common(k))
print(x)

print(type(x))

lst = []
lst.append(x[0][0])
lst.append((x[1][0]))

print(lst)




```
