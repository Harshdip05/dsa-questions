```python
names = ["Mary","John","Emma"]
heights = [180,165,170]

heights.sort()
heights.reverse()
print(heights)
```

```python
names = ["Mary","John","Emma"]
heights = [180,165,170]

# combine both
pairs = list(zip(heights, names))
print(pairs)

# sort by height descending
pairs.sort(reverse=True)
print(pairs)

# extract names
result = [name for height, name in pairs]

print(result)
```
